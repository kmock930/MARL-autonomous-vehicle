```python
# Use the options dict in the reset method
# This initialises the agent in location (0,0) and the goal in location (7,7)
env_8x8 = gym.make('SimpleGrid-8x8-v0', render_mode='human')
obs, info = env_8x8.reset(options={'start_loc':0, 'goal_loc':63})
```
